# Standardize the Variables

Time to standardize the variables.

** Import StandardScaler from Scikit learn.**

```python
from sklearn.preprocessing import StandardScaler
```

** Create a StandardScaler() object called scaler.**

```python
ss =StandardScaler()
```

** Fit scaler to the features.**

```python
ss.fit(df.drop('TARGET CLASS',axis=1))
```

**Use the .transform() method to transform the features to a scaled version.**

```python
scaled = ss.transform(df.drop('TARGET CLASS',axis=1))
```

**Convert the scaled features to a dataframe and check the head of this dataframe to make sure the scaling worked.**

```python
df_scaled = pd.DataFrame(scaled,columns = df.columns[:-1])
df_scaled.head(5)
```

# Train Test Split

**Use train_test_split to split your data into a training set and a testing set.**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_scaled,
                                                 df['TARGET CLASS'],
                                                 test_size=0.3)
```

# Using KNN

**Import KNeighborsClassifier from scikit learn.**

**Create a KNN model instance with n_neighbors=1**

```python
knn = KNeighborsClassifier(1)
```

**Fit this KNN model to the training data.**

```python
knn.fit(X_train,y_train)
```

# Predictions and Evaluations
Let's evaluate our KNN model!

**Use the predict method to predict values using your KNN model and X_test.**

```python
y_pred = knn.predict(X_test)
```

** Create a confusion matrix and classification report.**

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

** Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list. Refer to the lecture if you are confused on this step.**

```python
import matplotlib.pyplot as plt
np.sum(y_test.values!=y_pred)
```

```python
y_pred.shape
```

```python
errors = []
for i in range(1,45):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    errors.append(np.mean(y_test.values!=y_pred))
    
plt.figure(figsize=(15,8))
sns.lineplot(range(1,45),errors,marker='o',markerfacecolor='red',markersize=10)
plt.xlabel('K')
plt.ylabel('Error Rate')
```
